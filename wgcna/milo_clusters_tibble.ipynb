```r
library(limma)
```

```r
eset = ME[2:7] %>% t
group = ME$group
design = model.matrix(~0+group)
colnames(design) = levels(group)
corfit <- duplicateCorrelation(eset,design,block=ME$hash.mcl.ID)
```

```r
group %>% unique
```

```r
corfit$cor
```

```r
fit <- lmFit(eset,design,block=ME$hash.mcl.ID,correlation=corfit$consensus)
```

```r
cm = makeContrasts(
    contrasts=c("Day5.obob.FGF1vsVehPF = FGF1.Day5.obob - Veh_PF.Day5.obob",
      "Day14.obob.FGF1vsVehPF = FGF1.Day14.obob - Veh_PF.Day14.obob",
      "Day514.obob.FGF1vsVehPF = (FGF1.Day5.obob + FGF1.Day14.obob)/2 - (Veh_PF.Day5.obob+Veh_PF.Day14.obob)/2"),
    levels=design
)
cm
```

```r
contrasts= tribble(
    ~"name", ~"expression",        
    "Day5.obob.FGF1vsVehPF", "FGF1.Day5.obob - Veh_PF.Day5.obob",
    "Day14.obob.FGF1vsVehPF", "FGF1.Day14.obob - Veh_PF.Day14.obob",
    "Day514.obob.FGF1vsVehPF", "(FGF1.Day5.obob + FGF1.Day14.obob)/2 - (Veh_PF.Day5.obob+Veh_PF.Day14.obob)/2") %>%
deframe
contrasts
```

```r
cm = makeContrasts(
    contrasts=contrasts,
    levels=design
)
colnames(cm) = names(contrasts)
cm
```

```r
eset %>% head
```

```r
fit2 <- contrasts.fit(fit, cm)
```

```r
fit2 <- eBayes(fit2)
```

```r
topTable(fit2, coef = "Day5.obob.FGF1vsVehPF") %>%
mutate(significant = case_when(adj.P.Val < 0.05 ~ TRUE,
                               adj.P.Val >= 0.05 ~ FALSE))
```

```r
topTable(fit2, coef='Day514.obob.FGF1vsVehPF') %>%
mutate(significant = case_when(adj.P.Val < 0.05 ~ TRUE,
                               adj.P.Val >= 0.05 ~ FALSE))
```

```r
topTable(fit2, coef='Day14.obob.FGF1vsVehPF') %>%
mutate(significant = case_when(adj.P.Val < 0.05 ~ TRUE,
                               adj.P.Val >= 0.05 ~ FALSE))
```
