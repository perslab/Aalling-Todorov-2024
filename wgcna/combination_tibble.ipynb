```r
library(tidyverse)
library(targets)
library(tarchetypes)
```

```r
milo_clusters_tibble = qs::qread('milo_clusters_tibble.qs')
milo_clusters_tibble %>% head
```

```r
combination_recipe_00 = milo_clusters_tibble %>%
group_by(day) %>%
mutate(obj_names = paste0('gost_tt_summary_', label)) %>%
select(day, obj_names) %>%
mutate(day = paste0('combined_gost_tt_summary_', day)) %>%
rename(output_name = day) %>%
group_by(output_name) %>%
mutate(obj_names = list(obj_names)) %>%
ungroup %>%
distinct

combination_recipe_01 = milo_clusters_tibble %>%
group_by(day) %>%
mutate(obj_names = paste0('top_table_', label)) %>%
select(day, obj_names) %>%
mutate(day = paste0('combined_top_table_', day)) %>%
rename(output_name = day) %>%
group_by(output_name) %>%
mutate(obj_names = list(obj_names)) %>%
ungroup %>%
distinct

combination_recipe_02 = milo_clusters_tibble %>%
group_by(day) %>%
mutate(obj_names = paste0('hubgenes_summary_', label)) %>%
select(day, obj_names) %>%
mutate(day = paste0('combined_hubgenes_summary_', day)) %>%
rename(output_name = day) %>%
group_by(output_name) %>%
mutate(obj_names = list(obj_names)) %>%
ungroup %>%
distinct

combination_recipe_03 = milo_clusters_tibble %>%
group_by(day) %>%
mutate(obj_names = paste0('hubgenes_summary_tt_', label)) %>%
select(day, obj_names) %>%
mutate(day = paste0('combined_hubgenes_summary_tt_', day)) %>%
rename(output_name = day) %>%
group_by(output_name) %>%
mutate(obj_names = list(obj_names)) %>%
ungroup %>%
distinct

combination_recipe = rbind(combination_recipe_00, combination_recipe_01, combination_recipe_02, combination_recipe_03)


qs::qsave(combination_recipe, 'combination_recipe.qs')
combination_recipe
```
