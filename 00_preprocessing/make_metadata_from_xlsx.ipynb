```python
import pandas as pd
import os
from pathlib import Path
```

```python
project_dir = os.getenv('PROJECT_DIR')
metadata_path = Path(project_dir, 'data/received/metadata/0189_scnRNA_metadata_NA220620.xlsx')
metadata_path
```

```python
meta = pd.read_excel(metadata_path)
meta.head()
```

```python
keep_cols = ['Sample name', 'Strain', 'Condition 1', 'Condition 2', 'Collection date', 'Isolation date/pool', 'Comments']
meta = meta[keep_cols]
renamed_cols = {name:name.replace(' ', '_').lower() for name in keep_cols}
meta = meta.rename(columns=renamed_cols)
meta = meta.rename(columns={'condition_1': 'treatment',
                            'condition_2': 'time',
                            'isolation_date/pool': 'isolation_date',
                            'comments': 'batch'})
meta.head()
```

```python
collection_replace = {x: "20{}.{}.{}".format(str(x)[0:2], str(x)[2:4], str(x)[4:6]) \
                      for x in meta['collection_date'].unique()}
date_replace = {'22/6/2022': '2022.06.22',
                '23/6/2022': '2022.06.23',
                '24/6/2022': '2022.06.24'}
val_replace = {**collection_replace, **date_replace}
meta.replace(val_replace)
```

```python
metadata_new_path = Path(project_dir, 'data/meta/metadata.csv')
os.makedirs(metadata_new_path.parent, exist_ok=True)
```

```python
meta.to_csv(metadata_new_path, index=False)
```
