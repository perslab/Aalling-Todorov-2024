```r
merged = qs::qread('/projects/petar/fgf1/00_preprocessing_cellbender/merged_objects_nadia_fgf1_v2.qs')

obj_scop = qs::qread('/projects/petar/fgf1/00_preprocessing/_targets/objects/exp_04')
```

```r
merged_meta = merged %>% `[[` %>%
rownames_to_column %>%
separate(rowname, into = c("idx", "barcode"), sep = "_", remove=FALSE) %>%
select(-idx, -orig.ident)

meta_scop = obj_scop %>% `[[` %>%
rownames_to_column %>%
separate(rowname, into = c("barcode", "run_id1", "run_id2"), sep = "_", remove=FALSE)

merge_cellbender_scop = merged_meta %>%
left_join(meta_scop, by = c("Index.10x", "barcode"), suffix = c('.cellbender', '.scop')) %>%
select(nCount_RNA.cellbender, nCount_RNA.scop) %>%
mutate(diff = nCount_RNA.cellbender - nCount_RNA.scop) %>%
arrange(desc(diff)) %>%
mutate(diff_frac_cellbender = diff/nCount_RNA.cellbender)

merge_cellbender_scop %>%
pull(diff) %>%
quantile(probs = c(0, 0.25, 0.50, 0.75, 1)) %>% print
```
