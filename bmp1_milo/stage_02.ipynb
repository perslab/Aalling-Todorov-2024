```r
library(tidyverse)
library(targets)
library(tarchetypes)
```

```r
contrasts_recipe = qs::qread('da_recipe_bmp1.qs')
head(contrasts_recipe)
```

```r
# da_results_01_Sst_Pthlh.sc21___obobBL6
combination_recipe = contrasts_recipe %>%
select(label, day, cluster, contrast_name, suffix) %>% 
rowwise %>%
mutate(da_results_name = paste0("da_results_01_", label)) %>%
select(-label, -cluster) %>%
group_by(day, contrast_name, suffix) %>%
mutate(da_names = list(da_results_name)) %>%
select(-da_results_name) %>%
distinct %>%
mutate(output_suffix = paste0(day, '.', contrast_name)) %>%
mutate(output_name = paste0("combined_da_results", suffix, '_', output_suffix)) %>%
ungroup %>%
relocate(output_name, output_suffix)
```

```r
qs::qsave(combination_recipe, 'combination_recipe_bmp1.qs')
```
