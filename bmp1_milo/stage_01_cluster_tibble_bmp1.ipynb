```r
library(tidyverse)
library(targets)
```

```r
clusters_tibble = qs::qread('../01_milo/clusters_tibble.qs')
clusters_tibble %>% head
```

```r
clusters_tibble = clusters_tibble %>%
filter(day != 'Day14') %>%
filter(object == "exp_labelled_other") %>%
select(-object)

suffixes = c('_unimod', '_sct', '_sct_forb', '_sct_batchvar_all', '_sct_batchvar_forb', '_bmp_only')
input_obj_tibble = tibble(object = rlang::syms(c('bmp1_fgf1_mapped', 'bmp1_fgf1_mapped', 'bmp1_fgf1_mapped', 'bmp1_fgf1_mapped', 'bmp1_fgf1_mapped', 'bmp1_only_mapped')),
                          suffix = suffixes)
clusters_tibble = tidyr::crossing(clusters_tibble, input_obj_tibble) %>%
    mutate(label = paste0(label, suffix)) %>%
    mutate(seurat_obj = rlang::syms(paste0('obj_', label)))


for (sf in suffixes){
    clusters_tibble %>% filter(suffix == sf) %>% qs::qsave(., paste0('clusters_tibble_bmp1', sf, '.qs'))
}

for (sf in suffixes){
    qs::qread(paste0('clusters_tibble_bmp1', sf, '.qs')) %>% dim %>% print
}

clusters_tibble %>% qs::qsave('clusters_tibble_bmp1.qs')
clusters_tibble
```
