```r
library(tidyverse)
library(targets)
library(tarchetypes)
```

```r
milo_clusters_tibble = qs::qread('milo_clusters_tibble_bmp1.qs')
milo_clusters_tibble %>% head
```

```r
contrasts_recipe = qs::qread('../bmp1_milo/da_recipe_bmp1.qs')
head(contrasts_recipe)
```

```r
combination_recipe = contrasts_recipe %>%
tidyr::crossing(tibble(result_prefix = c('gost_tt_summary', 'top_table', 'hubgenes_summary', 'hubgenes_summary_tt'))) %>%
select(label, day, cluster, contrast_name, suffix, result_prefix) %>%
group_by(result_prefix, suffix) %>%
mutate(combined_name = paste0(result_prefix, suffix, day)) %>%
mutate(obj_name = paste0(result_prefix, '_', label, '_', day)) %>%
ungroup %>%
group_by(combined_name, suffix) %>%
mutate(obj_list = list(obj_name)) %>%
ungroup %>%
select(combined_name, obj_list) %>%
distinct

qs::qsave(combination_recipe, 'combination_recipe.qs')
combination_recipe
```
