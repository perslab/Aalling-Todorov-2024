```r
tar_load(combined_da_results_obob14v14,)
```

```r
tar_load(combined_da_results_all_obob14v14)
all = combined_da_results_all_obob14v14  %>%
filter(SpatialFDR < 0.1) %>%
group_by(labels) %>%
summarise(n_all = n())

tar_load(combined_da_results_Day14_obob14v14)
limited = combined_da_results_Day14_obob14v14  %>%
filter(SpatialFDR < 0.1) %>%
group_by(labels) %>%
summarise(n_limited = n())

full_join(all, limited) %>% 
mutate(diff = n_all - n_limited) %>%
arrange(desc(diff)) %>%
print(n=100)
```

```r
tar_load(combined_da_results_all_obob5v5)
all = combined_da_results_all_obob5v5  %>%
filter(SpatialFDR < 0.1) %>%
group_by(labels) %>%
summarise(n_all = n())

tar_load(combined_da_results_Day5_obob5v5)
limited = combined_da_results_Day5_obob5v5  %>%
filter(SpatialFDR < 0.1) %>%
group_by(labels) %>%
mutate(n_limited = n()) %>%
select(labels, n_limited, n)

full_join(all, limited) %>% 
mutate(diff = n_all - n_limited) %>%
arrange(desc(diff)) %>%
print(n=100)
```
