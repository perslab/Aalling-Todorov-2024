```r
library(tidyverse)
library(cxhull)
```

```r
points <- rbind(
  c(0.5, 0.5, 0.5),
  c(0, 0, 0),
  c(0, 0, 1),
  c(0, 1, 0),
  c(0, 1, 1),
  c(1, 0, 0),
  c(1, 0, 1),
  c(1, 1, 0),
  c(1, 1, 1)
)
hull <- cxhull(points)
```

```r
hull$vertices
```

```r
vtx = hull %>%
`[[`('vertices') %>%
data.frame

vtx
```

```r
hull$vertices %>% data.frame
```

```r
vtx = hull$vertices %>% 
data.frame %>% 
t  %>% 
data.frame %>% 
rownames_to_column %>%
filter(str_detect(rowname, "point")) %>%
select(-rowname) %>%
distinct

colnames(vtx) = c("x", 'y', 'z')

vtx
```
