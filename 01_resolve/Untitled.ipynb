```r
library(Seurat)
library(tidyverse)
library(ggplot2)
library(patchwork)
library(forcats)
library(Polychrome)
library(grid)
```

```r
xenium.obj = qs::qread('230502_xenium_sct_unimapped_cca.qs')
xenium.obj
```

```r
options(repr.plot.width=20, repr.plot.height=10)
ImageDimPlot(xenium.obj, fov = "fov", molecules = c("Pomc"), nmols = 20000, axes=TRUE)
```

```r
options(repr.plot.width=20, repr.plot.height=10)
ImageDimPlot(xenium.obj, fov = "fov", molecules = c("Pomc"), nmols = 20000, axes=TRUE)
```

```r
cropped.coords <- Crop(xenium.obj[["fov"]], x = c(5000, 12750), y = c(0, 6000), coords = "plot")
xenium.obj[["zoom.fov.arc"]] <- cropped.coords
DefaultBoundary(xenium.obj[["zoom.fov.arc"]]) <- "segmentation"
ImageDimPlot(xenium.obj, fov = "zoom.fov.arc", axes = TRUE, border.color = "white", border.size = 0.1, cols = "polychrome",
             coord.fixed = TRUE, 
             molecules = c("Pomc"),
             nmols = Inf) + theme(legend.position = "none")
```
