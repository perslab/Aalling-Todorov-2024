```r
xe_obj = qs::qread('_targets/objects/obj_other_02')
obj_fgf1 = qs::qread('_targets/objects/obj_d5_other')
selected_label = 'OPC'


transfer_data_rctd_polar_label = function(xe_obj, obj_fgf1, selected_label) {
  tryCatch(
    {
      xe_selected_cells = xe_obj %>% `[[` %>% filter(labels == selected_label) %>% rownames
      xe_obj = xe_obj %>%
        subset(cells = xe_selected_cells)
      fgf1_selected_cells = obj_fgf1 %>% `[[` %>% filter(labels == selected_label) %>% rownames
      obj_fgf1 = obj_fgf1 %>%
        subset(cells = fgf1_selected_cells)
      annot_df = run_rctd(xe_obj, obj_fgf1, 'polar_label', confidence_threshold = 5, doublet_threshold=20)
      annot_df
    },
    error = function(e) {
      message("An error occurred: ", e$message)
      return(e)
    }
  )
}
```

```r

transfer_data_rctd_polar_label(xe_obj, obj_fgf1, selected_label)
```
