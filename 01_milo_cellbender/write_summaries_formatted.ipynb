```r
split_data_to_excel <- function(input_data, output_dir) {
  # group the data by the "tag" column and create a list of data frames
  grouped_data <- input_data %>% group_split(tag)

  # loop through the groups and write each one to a separate Excel file in a subfolder for each combination of values
  walk2(grouped_data, unique(input_data$tag), function(df, group_val) {
    folder_name <- paste0(output_dir, "/", df$fgf1_day[1], ".", df$fgf1_comparison[1], "__v__", df$bl6_day[1], ".", df$bl6_comparison[1])
    if (!dir.exists(folder_name)) {
      dir.create(folder_name, recursive = TRUE)
    }
    write_xlsx(df, paste0(folder_name, "/", group_val, ".xlsx"))
  })
}
```

```r
input_dir <- "outputs/plots/deg_summary/"
output_dir <- "outputs/plots_seurat/"

# create the output directory if it does not exist
if (!dir.exists(output_dir)) {
  dir.create(output_dir, recursive = TRUE)
}

# define the regular expression to extract the relevant parts of the tag
tag_regex <- "___(\\w+)\\.?(\\w+)?__v__(\\w+)\\.?(\\w+)?___"

# loop through the PDF files in the input directory and copy them to the output directory with the new subfolder structure
pdf_files <- list.files(input_dir, pattern = "*.pdf", full.names = TRUE)
for (pdf_file in pdf_files) {
  file_name <- basename(pdf_file)
  tag_match <- str_match(file_name, tag_regex)
  fgf1_day <- tag_match[[2]]
  fgf1_comparison <- tag_match[[3]]
  bl6_day <- tag_match[[4]]
  bl6_comparison <- tag_match[[5]]
  subfolder_name <- paste0(fgf1_day, ".", fgf1_comparison, "__v__", bl6_day, ".", bl6_comparison)
  subfolder_path <- file.path(output_dir, subfolder_name)
  if (!dir.exists(subfolder_path)) {
    dir.create(subfolder_path, recursive = TRUE)
  }
  new_file_path <- file.path(subfolder_path, file_name)
  file.copy(pdf_file, new_file_path)
}
```

```r
all_restored_summary = qs::qread('_targets/objects/all_restored_summary')
# Split the tibble by the 'comparison' column
split_data <- split(all_restored_summary, all_restored_summary$comparison)

dir.create("outputs/restored_summaries", recursive = TRUE)

# Write out a file for each split
walk(split_data, ~ {
  comparison_name <- unique(.x$comparison)
  file_name <- paste0("outputs/restored_summaries/restored_summary_", comparison_name, ".xlsx")
  write_xlsx(.x, file_name)
})
```

```r
process_combined_da_results = function(da_results_all){
    da_summary = da_results_all %>%
    group_by(labels, polarity) %>%
    summarise(n = n()) %>%
    arrange(desc(labels))

    da_label_groups = da_results_all %>%
    group_by(labels) %>%
    summarise(n_group = n()) %>%
    arrange(desc(labels))

    da_summary = merge(da_summary, da_label_groups)

    da_summary = da_summary %>% mutate(pct_group = n/n_group)

    da_summary_sum = da_summary %>% 
      filter(polarity == 'none') %>%
      mutate(pct_group_changed = (n_group - n)/n_group) %>%
      select(-c(polarity, pct_group, n)) %>%
      arrange(desc(pct_group_changed), desc(n_group))

    da_results_all$labels = da_summary_sum %>% arrange(pct_group_changed, n_group) %>% pull(labels) %>% factor(da_results_all$labels, levels=.)
    da_results_all
}

input_dir <- "_targets/objects"
output_dir <- "outputs/milo_DAbeeswarms/"
dir.create(output_dir, recursive = TRUE)
da_results_files <- list.files(input_dir, pattern = "combined_da_results", full.names = TRUE)
for (da_results_file in da_results_files) {
    file_name = basename(da_results_file)
    da_results = qs::qread(da_results_file) %>% process_combined_da_results
    my_plot = miloR::plotDAbeeswarm(da_results, group.by='labels')
    plot_path = paste0(output_dir, "DAbeeswarm_", file_name, '.pdf')
    ggsave(plot_path, plot = my_plot, width = 10, height = 30)
}
```
