```r
library(tidyverse)
```

```r
stage_05 = qs::qread('annotate_nh_grouping_recipe.qs')
stage_05
```

```r
stage_xx = stage_05 %>%
distinct(da_results_nhg_output_suffix, comparison) %>%
rowwise %>%
mutate(da_results_nhg_name = paste0("da_results_nhg_idx_", da_results_nhg_output_suffix)) %>%
group_by(comparison) %>%
mutate(da_results_nhg_names = list(da_results_nhg_name)) %>%
distinct(comparison, da_results_nhg_names) %>%
rowwise %>%
mutate(output_name = paste0('combined_da_results_nhg_', comparison)) %>%
ungroup

stage_xx %>% qs::qsave('stage_xx_recipe.qs')
stage_xx
```
