```r
library(tidyverse)
library(targets)
library(tarchetypes)
```

```r
contrasts_recipe_class = qs::qread('da_recipe_class.qs')
contrasts_recipe_lvl1 = qs::qread('da_recipe_lvl1.qs')
contrasts_recipe_labels_chunk = qs::qread('da_recipe_labels_chunk.qs')
contrasts_recipe = bind_rows(contrasts_recipe_class, contrasts_recipe_lvl1, contrasts_recipe_labels_chunk)
dim(contrasts_recipe)
head(contrasts_recipe)
```

```r
# da_results_01_Sst_Pthlh.sc21___obobBL6
combination_recipe = contrasts_recipe %>%
select(label, day, cluster, contrast_name) %>% 
rowwise %>%
mutate(da_results_name = paste0("da_results_01_", label)) %>%
mutate(da_results_name_idx = paste0("da_results_02_idx_", label)) %>%
select(-label, -cluster) %>%
group_by(day, contrast_name) %>%
mutate(da_names = list(da_results_name)) %>%
mutate(da_names_idx = list(da_results_name_idx)) %>%
select(-da_results_name, -da_results_name_idx) %>%
distinct %>%
mutate(output_suffix = paste0(day, '.', contrast_name)) %>%
mutate(output_name = paste0("combined_da_results_", output_suffix)) %>%
mutate(output_name_idx = paste0("combined_da_results_idx_", output_suffix)) %>%
ungroup %>%
relocate(output_name, output_name_idx, output_suffix)
```
