```r
library(tidyverse)
library(Seurat)
library(patchwork)
```

```r
markers = qs::qread('_targets_nhBulkDESeq/objects/deg_ensmus_Htr3b___Day5.obob5v5__v__Day5.obobBL6d5___fgf1_grouping.neg_vs_none')

selected_features = markers %>% 
filter(p_val_adj < 0.05) %>%
filter(avg_log2FC > 0) %>%
# mutate(gsea_sort_score = (-log10(.[[3]]))) %>%
# mutate(gsea_sort_score = .[[2]] * gsea_sort_score) %>%
# mutate(gsea_sort_score = sign(.[[2]]) * gsea_sort_score) %>%
# # arrange(desc(gsea_sort_score)) %>%
pull(GeneID)

markers %>% 
filter(p_val_adj < 0.05) %>%
filter(avg_log2FC > 0) %>%
mutate(gsea_sort_score = (-log10(.[[3]]))) %>%
mutate(gsea_sort_score = .[[2]] * gsea_sort_score) %>%
mutate(gsea_sort_score = sign(.[[2]]) * gsea_sort_score) %>%
arrange(desc(gsea_sort_score)) %>%
rownames %>% length

selected_features %>% paste0(collapse=' ')
```

```r
seurat_obj = qs::qread('_targets_MAST/objects/obj_Day5.Htr3b')
DefaultAssay(seurat_obj) = 'RNA'
options(repr.plot.width = 10, repr.plot.height = 20)
FeaturePlot(seurat_obj, features = selected_features %>% `[`(1:18), 
            slot = 'counts', order=TRUE, ncol=3)
```
