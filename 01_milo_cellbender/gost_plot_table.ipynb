```r
library(igraph)

make_grouped_gost = function(df){
    #uses tidyverse, igraph
    # Your original transformation
    df_long <- df %>% 
      tidyr::unnest(parents) %>%  # parents is a list column containing one vector of go terms per row
    #   rename(to=term_id, from=parents) %>% 
      filter((parents %in% term_id)) %>%
      rename(to=parents, from=term_id) %>% 
      select(to, from) %>%
      distinct()

    all_term_ids = data.frame(name = df %>% distinct(term_id) %>% pull(term_id))
    # Create a directed graph from the long format data frame
    g <- graph_from_data_frame(df_long, vertices=all_term_ids,  directed = TRUE)

    # Get all node names
    all_nodes <- V(g)$name

    # Find nodes with no incoming edges (i.e., no parents)
    only_parent_nodes <- all_nodes[degree(g, mode = "in") == 0]

    # Create an empty tibble to store our results
    results <- tibble(hub_node = character(), contained_nodes = list(), path=list(), terminal_nodes = list())

    # Iterate over each parent node
    for (hub_node in only_parent_nodes) {
      # Find all directed paths from the current node to all other nodes
      paths <- all_simple_paths(g, from = hub_node, mode = "out")  # Modified to find directed paths

      # Extract the downstream nodes from each path, flatten into a single vector, remove duplicates, and convert indices to GO terms
      downstream_nodes <- unique(unlist(paths))
      downstream_nodes <- get.vertex.attribute(g, "name", index = downstream_nodes)

      chain_of_nodes = downstream_nodes
      # Exclude the hub_node itself from the downstream_nodes
      downstream_nodes <- setdiff(downstream_nodes, hub_node)

      # Exclude parent nodes from downstream_nodes
      parent_nodes <- df_long$from[df_long$to == hub_node]
      downstream_nodes <- setdiff(downstream_nodes, parent_nodes)

      # Get terminal (leaf) nodes for each path
      terminal_nodes <- downstream_nodes[degree(g, downstream_nodes, mode = "out") == 0]

      # Add the current parent node and its downstream nodes to our results tibble
      results <- results %>% add_row(hub_node = hub_node, contained_nodes = list(downstream_nodes), path=list(chain_of_nodes), terminal_nodes=list(terminal_nodes))
    }

    results_joined <- results %>% 
      left_join(df, by = c("hub_node" = "term_id")) %>% 
      rename(hub_node_name = term_name) %>% 
      mutate(path_names = map(path, ~df$term_name[df$term_id %in% .x])) %>%
      mutate(terminal_names = map(terminal_nodes, ~df$term_name[df$term_id %in% .x]))

    # If you wish to remove the original 'contained_nodes' column you can use the select() function:
    # results <- results %>% select(-contained_nodes)

    grouped_gost = results_joined
    grouped_gost
}
```

```r
format_grouped_gost = function(grouped_gost){
    grouped_gost = grouped_gost %>%
        distinct(hub_node, .keep_all = TRUE) %>%
        select(tag, cluster, fgf1_day, fgf1_comparison, bl6_day, bl6_comparison, 
               grouping, cells_a, cells_b, 
               hub_node, hub_node_name, terminal_names, term_size, query_size, intersection_size, p_value, gene_ids) %>%
        rowwise %>%
    mutate(gene_ids = str_replace_all(gene_ids, ',', '; ')) %>%
    ungroup %>%
    rowwise %>%
    mutate(terminal_names = terminal_names %>% unlist %>% paste0(collapse = '; ')) %>%
    ungroup %>%
        arrange(p_value)
    grouped_gost
}
```
