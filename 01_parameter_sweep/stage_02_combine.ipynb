```r
library(tidyverse)
library(targets)
library(tarchetypes)
```

```r
clusters_tibble = qs::qread('clusters_tibble_lvl1.qs')

clusters_tibble$batchy_score_tibble = clusters_tibble %>%
    pull(label) %>%
    paste0("batchy_score_tibble_", .) %>%
    rlang::syms()
clusters_tibble 
```

```r
combination_recipe = clusters_tibble %>%
group_by(old.label) %>%
mutate(da_names_idx = list(batchy_score_tibble %>% as.character)) %>%
ungroup %>%
rowwise %>%
mutate(output_name = old.label) %>%
mutate(output_name = paste0('combined_batchy_', output_name)) %>%
mutate(combination_tibble_obj = output_name %>% rlang::syms()) %>%
ungroup %>%
select(output_name, combination_tibble_obj, old.label, da_names_idx) %>%
distinct

combination_recipe %>% ungroup %>% qs::qsave('combination_recipe.qs')
```
